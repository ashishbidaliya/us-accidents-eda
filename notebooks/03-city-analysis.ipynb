# City Analysis

In this section:
1. Analyze accident distribution across cities
2. Identify cities with highest accident counts
3. Understand accident frequency distribution

## Import Libraries

```python
import pandas as pd
import seaborn as sns

sns.set_style("darkgrid")
```

## Load Dataset

```python
df = pd.read_csv("../data/raw/US_Accidents_March23.csv")
```

## City Column

```python
df.City
```

## Unique Cities

```python
cities = df.City.unique()

len(cities)
```

## Accident Count By City

```python
cities_by_accident = df.City.value_counts()

cities_by_accident
```

## Top 20 Cities

```python
cities_by_accident[:20]
```

## Type Check

```python
type(cities_by_accident)
```

## Top 20 Cities Visualization

```python
cities_by_accident[:20].plot(kind='barh')
```

## Distribution Of Accidents Per City

```python
sns.histplot(
    cities_by_accident,
    log_scale=True
)
```

## Cities With Only One Accident

```python
cities_by_accident[
    cities_by_accident == 1
]
```
