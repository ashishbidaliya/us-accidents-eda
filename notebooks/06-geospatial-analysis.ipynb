# Geospatial Analysis

In this section:
1. Analyze accident locations
2. Visualize accident coordinates
3. Create accident heatmaps

## Import Libraries

```python
import pandas as pd
import seaborn as sns
import folium

from folium.plugins import HeatMap
```

## Load Dataset

```python
df = pd.read_csv(
    "../data/raw/US_Accidents_March23.csv",
    usecols=['Start_Lat', 'Start_Lng']
)
```

## Latitude Column

```python
df.Start_Lat
```

## Longitude Column

```python
df.Start_Lng
```

## Sample Dataset

```python
sample_df = df.sample(int(0.1 * len(df)))
```

## Scatter Plot

```python
sns.scatterplot(
    x=sample_df.Start_Lng,
    y=sample_df.Start_Lat,
    size=0.001
)
```

## First Latitude & Longitude

```python
lat, lon = df.Start_Lat[0], df.Start_Lng[0]

lat, lon
```

## Sample Latitude & Longitude Rows

```python
for _, row in df[['Start_Lat', 'Start_Lng']].sample(100).iterrows():
    print(row)
```

## Zip Latitude & Longitude

```python
zip(list(df.Start_Lat), list(df.Start_Lng))
```

## HeatMap Sample Dataset

```python
sample_df = df.sample(int(0.001 * len(df)))

lat_lon_pairs = list(
    zip(
        list(sample_df.Start_Lat),
        list(sample_df.Start_Lng)
    )
)
```

## HeatMap

```python
map = folium.Map()

HeatMap(lat_lon_pairs).add_to(map)

map
```
