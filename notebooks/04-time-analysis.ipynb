# Time Analysis


In this section:
1. Analyze accidents by hour
2. Analyze weekday vs weekend trends
3. Analyze yearly and monthly trends

## Import Libraries

```python
import pandas as pd
import seaborn as sns

sns.set_style("darkgrid")
```

## Load Dataset

```python
df = pd.read_csv(
    "../data/raw/US_Accidents_March23.csv",
    usecols=['Start_Time', 'Source']
)
```

## Start Time Column

```python
df.Start_Time
```

## Convert To Datetime

```python
df.Start_Time = pd.to_datetime(
    df.Start_Time,
    format='mixed'
)
```

Figure out how to show percentages.

## Accidents By Hour

```python
sns.histplot(
    df.Start_Time.dt.hour,
    bins=24
)
```

- A high percentage of accidents occur between 6 am to 10 am (probably people in a hurry to get to work)
- Next highest percentage is 3 pm to 6 pm.

## Accidents By Day Of Week

```python
sns.histplot(
    df.Start_Time.dt.dayofweek,
    bins=7
)
```

## Weekend vs Weekday Distribution

Is the distribution of accidents by hour the same on weekends as on weekdays?

## Sunday Accident Distribution

```python
sundays_start_time = df.Start_Time[
    df.Start_Time.dt.dayofweek == 6
]

sns.histplot(
    sundays_start_time.dt.hour,
    bins=24
)
```

## Monday Accident Distribution

```python
monday_start_time = df.Start_Time[
    df.Start_Time.dt.dayofweek == 0
]

sns.histplot(
    monday_start_time.dt.hour,
    bins=24
)
```

On Sundays, the peak occurs between 10 am and 3 pm, unlike weekdays.

## Monthly Trend Analysis

```python
df_2019 = df[
    df.Start_Time.dt.year == 2019
]

df_2019_Source1 = df_2019[
    df_2019.Source == 'Source1'
]

sns.histplot(
    df_2019_Source1.Start_Time.dt.month,
    bins=12
)
```
