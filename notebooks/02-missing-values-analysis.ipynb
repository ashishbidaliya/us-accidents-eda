# Missing Values Analysis

In this section:
1. Check missing values percentage
2. Identify columns with high missing values
3. Visualize missing data distribution

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("../data/raw/US_Accidents_March23.csv")
```

## Percentage Of Missing Values Per Column

```python
missing_percentages = (
    df.isna()
    .sum()
    .sort_values(ascending=False)
    / len(df)
)

missing_percentages
```

## Type Check

```python
type(missing_percentages)
```

## Missing Values Visualization

```python
missing_percentages[
    missing_percentages != 0
].plot(kind='barh')
```
